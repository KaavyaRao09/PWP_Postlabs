```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r'F:\College\Sem 3\PWP\MU.jpg')
img_array = np.array(img)
R = img_array[:, :, 0]
G = img_array[:, :, 1]
B = img_array[:, :, 2]
plt.figure(figsize=(10, 4))
plt.subplot(1, 4, 1)
plt.imshow(img_array)
plt.title("Original")
plt.axis('off')
plt.subplot(1, 4, 2)
plt.imshow(R, cmap='gray')
plt.title("R channel")
plt.axis('off')
plt.subplot(1, 4, 3)
plt.imshow(G, cmap='gray')
plt.title("G channel")
plt.axis('off')
plt.subplot(1, 4, 4)
plt.imshow(B, cmap='gray')
plt.title("B channel")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r'F:\College\Sem 3\PWP\MU.jpg')
img_array = np.array(img)

top = bottom = left = right = 20

H, W = img_array.shape[:2]
C = 1 if img_array.ndim == 2 else img_array.shape[2]

padded = np.zeros((H + top + bottom, W + left + right, C), dtype=img_array.dtype)
padded[top:top+H, left:left+W, ...] = img_array

plt.imshow(padded)
plt.axis('off')
plt.title("Padded Image (Black Border)")
plt.show()
```
