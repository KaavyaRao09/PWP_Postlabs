```python
import numpy as np
import matplotlib.pyplot as plt
# e. Reverse 5 Hz sine wave in time
original_signal_e = np.sin(2 * np.pi * 5 * t1)
reversed_signal_e = original_signal_e[::-1]
plt.subplot(3, 2, 5)
plt.plot(t1, original_signal_e, label="Original")
plt.plot(t1, reversed_signal_e, label="Time reversed", linestyle="--")
plt.title("e. Original and time-reversed 5 Hz sine wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# d. Scale 10 Hz sine wave amplitude by 3
original_signal_d = np.sin(2 * np.pi * 10 * t1)
scaled_signal_d = 3 * original_signal_d
plt.subplot(3, 2, 4)
plt.plot(t1, original_signal_d, label="Original")
plt.plot(t1, scaled_signal_d, label="Scaled by 3", linestyle="--")
plt.title("d. Original and amplitude-scaled 10 Hz sine wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# b. Multiply 5 Hz sine and 10 Hz cosine waves, sampled at 500 Hz for 2 seconds
fs2 = 500
t2 = np.linspace(0, 2, 2*fs2, endpoint=False)
sine_5Hz_b = np.sin(2 * np.pi * 5 * t2)
cosine_10Hz = np.cos(2 * np.pi * 10 * t2)
mult_signal = sine_5Hz_b * cosine_10Hz
plt.subplot(3, 2, 2)
plt.plot(t2, mult_signal)
plt.title("b. Element-wise product of 5 Hz sine and 10 Hz cosine")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# a. Add two sine waves (5 Hz and 10 Hz) sampled at 1000 Hz for 1 second
fs1 = 1000
t1 = np.linspace(0, 1, fs1, endpoint=False)
sine_5Hz = np.sin(2 * np.pi * 5 * t1)
sine_10Hz = np.sin(2 * np.pi * 10 * t1)
sum_signal = sine_5Hz + sine_10Hz
plt.figure(figsize=(12, 8))
plt.subplot(3, 2, 1)
plt.plot(t1, sum_signal)
plt.title("a. Sum of 5 Hz and 10 Hz sine waves")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
```
